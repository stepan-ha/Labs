# Імпорт бібліотек

---



```python
import numpy as np
import matplotlib.pyplot as plt
```

# Ініціалізація Змінних

---



```python
S0 = 100
mu = 0.1
sigma = 0.05
T = 1
N = 1000
dt = T / N
t = np.linspace(0, T, N)

np.random.seed(42)
W = np.random.normal(0, np.sqrt(dt), N).cumsum()

S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)
```

# Перший Графік

---



```python
plt.figure(figsize=(10, 6))
plt.plot(t, S, label='Аналітичний розв\'язок')
plt.title('Аналітичний розв\'язок для геометричного броунівського руху')
plt.xlabel('Час')
plt.ylabel('Ціна акції')
plt.legend()
plt.grid(True)
plt.show()
```

# Фінкції обрахунку відхилення

---



```python
def euler_maruyama_error(h, N, mu, sigma, S0, T):
    dt = T / N
    errors = []
    for _ in range(50):
        W = np.random.normal(0, np.sqrt(dt), N).cumsum()
        S_analytical = S0 * np.exp((mu - 0.5 * sigma**2) * np.linspace(0, T, N) + sigma * W)

        S_euler = np.zeros(N)
        S_euler[0] = S0
        for i in range(1, N):
            S_euler[i] = S_euler[i-1] + mu * S_euler[i-1] * dt + sigma * S_euler[i-1] * (W[i] - W[i-1])

        errors.append(np.mean(np.abs(S_analytical - S_euler)))
    return np.mean(errors)

def milstein_error(h, N, mu, sigma, S0, T):
    dt = T / N
    errors = []
    for _ in range(50):
        W = np.random.normal(0, np.sqrt(dt), N).cumsum()
        S_analytical = S0 * np.exp((mu - 0.5 * sigma**2) * np.linspace(0, T, N) + sigma * W)

        S_milstein = np.zeros(N)
        S_milstein[0] = S0
        for i in range(1, N):
            dW = W[i] - W[i-1]
            S_milstein[i] = S_milstein[i-1] + mu * S_milstein[i-1] * dt + sigma * S_milstein[i-1] * dW + 0.5 * sigma**2 * S_milstein[i-1] * (dW**2 - dt)

        errors.append(np.mean(np.abs(S_analytical - S_milstein)))
    return np.mean(errors)
```

# **Розв'язок**

---



```python
np.random.seed(42)
W = np.random.normal(0, np.sqrt(dt), N).cumsum()

S_analytical = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

S_euler = np.zeros(N)
S_euler[0] = S0
for i in range(1, N):
    S_euler[i] = S_euler[i-1] + mu * S_euler[i-1] * dt + sigma * S_euler[i-1] * (W[i] - W[i-1])

S_milstein = np.zeros(N)
S_milstein[0] = S0
for i in range(1, N):
    dW = W[i] - W[i-1]
    S_milstein[i] = S_milstein[i-1] + mu * S_milstein[i-1] * dt + sigma * S_milstein[i-1] * dW + 0.5 * sigma**2 * S_milstein[i-1] * (dW**2 - dt)

plt.figure(figsize=(10, 6))
plt.plot(t, S_analytical, label='Аналітичний розв\'язок')
plt.plot(t, S_euler, label='Метод Ейлера', linestyle='--')
plt.plot(t, S_milstein, label='Метод Мільштейна', linestyle=':')
plt.title('Порівняння аналітичного розв\'язку та числових методів')
plt.xlabel('Час')
plt.ylabel('Ціна акції')
plt.legend()
plt.grid(True)
plt.show()
```

# **Обрахунок похибок**

---



```python
h_values = [0.01, 0.005, 0.0025]
N_values = [int(T / h) for h in h_values]

euler_errors = [euler_maruyama_error(h, N, mu, sigma, S0, T) for h, N in zip(h_values, N_values)]
milstein_errors = [milstein_error(h, N, mu, sigma, S0, T) for h, N in zip(h_values, N_values)]

for i in range(0, len(h_values)):
    print(f"h = {h_values[i]}, Ейлер на попередньому кроці: {euler_errors[i-1] / euler_errors[i]}")
    print(f"h = {h_values[i]}, Мільштейн на попередньому кроці: {milstein_errors[i-1] / milstein_errors[i]}")
```
